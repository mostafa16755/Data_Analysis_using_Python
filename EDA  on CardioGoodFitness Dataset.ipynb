## Insight.4)Sales per Ages

```python
def rep(age):
    if(age <= 35):
        return 'age less than or equel 35'
    if(age > 35):
        return 'age more than 35'
```

```python
df['Age_Bracek'] = df['Age'].apply(rep)
```

```python
sns.countplot(y='Product', data=df, hue='Age_Bracek');
```
