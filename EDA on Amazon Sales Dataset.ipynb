```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading dataset
data = pd.read_csv('C:/Users/Mostafa/Data/amazon.csv')
```

```python
#check first five rows
data.head()
```

```python
#Check columns names
data.columns
```

```python
#remove some columns
data.drop(columns = ['user_id', 'user_name', 'review_id', 'review_title', 'review_content', 'product_id'], inplace = True)
```

```python
#check number of rows and columns
data.shape
```

```python
data.info()
```

```python
#check for nulls
data.isna().values.any()
```

```python
#double check for details of nulls
data.isnull().sum()
```

```python
#remove rows that have nulls
data.dropna(inplace = True)
```

```python
#check for dublicated values
data.duplicated().sum()
```

```python
#change data type
data['discounted_price'] = data['discounted_price'].str.replace('₹', '')
data['discounted_price'] = data['discounted_price'].str.replace(',', '').astype(float)
data['actual_price'] = data['actual_price'].str.replace('₹', '')
data['actual_price'] = data['actual_price'].str.replace(',', '').astype(float)
data['discount_percentage'] = data['discount_percentage'].str.replace('%', '').astype(float)
```

```python
#check unique values in 'rating' column
data['rating'].unique()
```

```python
data.loc[data['rating'] == '|']
```

```python
#drop bad value
data.drop(index = 1279, inplace = True)
```

```python
data['rating'] = data['rating'].astype(float)
```

```python
#check for summary statistic
data.describe()
```

### Data Visualization

```python
#find the correlation of dataset
sns.heatmap(data.corr(), annot=True);
```

```python
data['category'] = data.category.apply(lambda x: x.split('|')[0])
```

### 1) Sales per Category

```python
data['category'].value_counts().plot(
    kind = 'bar',
    title = 'Sales per Category',
    xlabel = 'Category',
    ylabel = 'Number of sales transactions',
    color = 'limegreen');
```

### 2) Sales per Discount Percentage

```python
data['discount_percentage'].value_counts().head(35).plot(
    kind = 'bar',
    title = 'Sales per Discount Percentage',
    xlabel = 'Discount Percentage',
    ylabel = 'Number of sales transactions',
    color = 'limegreen');
```

### 3) Sales per Rating

```python
data['rating'].value_counts().head(40).plot(
    kind = 'bar',
    title = 'Sales per Rating',
    xlabel = 'Rate',
    ylabel = 'Number of sales transactions',
    color = 'limegreen');
```

### 4) Sales per Price

```python
data['actual_price'].value_counts().head(35).plot(
    kind = 'bar',
    title = 'Sales per Price',
    xlabel = 'Rate',
    ylabel = 'Number of sales transactions',
    color = 'limegreen');
```
